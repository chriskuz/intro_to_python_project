```python
day_of_week_hierarchy_dict = {
    "Sunday":1,
    "Monday":2,
    "Tuesday":3, 
    "Wednesday":4, 
    "Thursday":5, 
    "Friday":6,
    "Saturday":7 
}

df["searchDayOfWeekHierarchyEncoded"] = df["searchDayOfWeek"].map(day_of_week_hierarchy_dict)
df["searchDayOfWeekSine"] = np.sin(2 * np.pi * df["searchDayOfWeekHierarchyEncoded"] / 7)
df["searchDayOfWeekCosinee"] = np.cos(2 * np.pi * df["searchDayOfWeekHierarchyEncoded"] / 7)

testing_df = df.copy()

unique_days_df = testing_df.drop_duplicates(subset="searchDayOfWeekHierarchyEncoded").reset_index()

viewable_filter = [
    "searchDayOfWeek", 
    "searchDayOfWeekHierarchyEncoded", 
    "searchDayOfWeekSine",
    "searchDayOfWeekCosinee"   
]

unique_days_df[viewable_filter]
```
