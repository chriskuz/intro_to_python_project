# Rapid API Testing for Trip Advisor

## Data Prep Information

```python
with open("./secrets.json") as f:
    information = json.load(f)
```

```python
host_name = information.get("tripadvisor_api_host")
api_key = information.get("tripadvisor_api_key")
```

## Search Flight MultiCity

```python
url = "https://tripadvisor16.p.rapidapi.com/api/v1/flights/searchFlightsMultiCity"

querystring = {"legs":[{'sourceAirportCode':'BOS','destinationAirportCode':'LON','date':'2023-10-18'},{'sourceAirportCode':'LON','destinationAirportCode':'BOS','date':'2023-10-26'}],"classOfService":"ECONOMY","sortOrder":"ML_BEST_VALUE","currencyCode":"USD"}

headers = {
	"x-rapidapi-key": api_key,
	"x-rapidapi-host": host_name
}

response = requests.get(url, headers=headers, params=querystring)

multicity_json = response.json()
```

## Search Flight

```python
import requests

url = "https://tripadvisor16.p.rapidapi.com/api/v1/flights/searchFlights"

querystring = {"sourceAirportCode":"BOM","destinationAirportCode":"DEL","date":"2024-10-26","itineraryType":"ONE_WAY","sortOrder":"ML_BEST_VALUE","numAdults":"1","numSeniors":"0","classOfService":"ECONOMY","pageNumber":"1","nearby":"yes","nonstop":"yes","currencyCode":"USD","region":"USA"}

headers = {
	"x-rapidapi-key": api_key,
	"x-rapidapi-host": host_name
}

response = requests.get(url, headers=headers, params=querystring)

flight_json = response.json()
```
